```python
import reader
path = '/nfs/OPTI/bzfander/lrs_python/data/MITB_889_boxed.ine'
parsed = reader.reader(path)
```

```python
from bland import Bland
from crisscross import CrissCross
lrs = Bland(*parsed)
lrs.augment_matrix_with_objective()
lrs.init_dicts()
lrs.add_box_constraints(lrs.bounding_box)
```

```python
lrs.first_basis()
```

```python
search = lrs.search()
from lrs import SearchStatus
status = SearchStatus.NONE
i = 0
incremented = 0
backtracked = 0
newbasis = 0
```

```python
import logging
logger = logging.getLogger()
logger.setLevel(logging.INFO)
import sys
logging.disable(sys.maxsize)
```

```python
for i, status in enumerate(search):
    
    if status == SearchStatus.INCREMENTED:
        incremented += 1
    elif status == SearchStatus.BACKTRACKED:
        backtracked += 1
    elif status == SearchStatus.NEWBASIS:
        newbasis += 1
    elif status == SearchStatus.DONE or i == 10000:
        break
```
